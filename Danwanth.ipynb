```python
import math
s=int(input("Enter the no of elements: "))
l=[]
for i in range(s):
    k=int(input("Enter the vector component:"))
    l.append(k)
sum=0
for i in l:
    sum+=i**2

print("The L2 norm is: ",math.sqrt(sum))
```

```python
import math
s=int(input("Enter the dimensions of vector: "))
l1,l2=[],[]
for i in range(s):
    k=int(input("Enter the vector component of u:"))
    l1.append(k)

for i in range(s):
    k=int(input("Enter the vector component of v:"))
    l2.append(k)

sum=0
for j in range(s):
    sum+=l1[j]*l2[j]

if(sum==0):
    print("orthogonal")
else:
    print("not")
```

```python
s=int(input("Enter the no of elements: "))
l=[]
for i in range(s):
    k=int(input("Enter the element:"))
    l.append(k)
l.sort()

def median():
  mid=s//2

  if s%2 == 0:
    return ((l[mid-1]+l[mid])/2)
  else:
    return l[mid]

m=median()
print(m)




```

```python
c=int(input("Enter the no of columns: "))
r=int(input("Enter the no of rows for matrix: "))
l=[]
d=[]
for i in range(r):
    p=[]
    for j in range(c):
      h=int(input("Enter the "+str(i+1)+"th row "+str(j+1)+"th column: "))
      p.append(h)
    l.append(p)
for j in range(c):
  h=int(input("Enter the row matrix element"+str(j+1)+":"))
  d.append(h)
print(l)
print(d)
l1=[]
for i in range(r):
  sumv=0
  for j in range(c):
    sumv+=d[j]*l[i][j]
  l1.append(sumv)

print("the product of row vector is: ",l1)

```

```python
m=int(input("Enter the mean: "))
sd=int(input("Enter the standard deviation: "))
x=int(input("Enter the x value: "))
def zscore(m,sd,x):
  z=(x-m)/sd
  return z

zs=zscore(m,sd,x)
print("The z-score normalisation of "+str(x)+" is : "+str(zs))
```
